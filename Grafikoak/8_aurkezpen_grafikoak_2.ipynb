```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# Funtzioak definitu

x = np.linspace(-4,4,100)
def f(x):
    return x**2
def df(x):
    return 2*x
y = f(x)

def gd(f,df,x0,nu,T):
    v = np.zeros(T)
    v[0] = x0
    for i in range(1, T):
        v[i] = v[i-1] - nu * df(v[i-1])
    return v, v[-1]
        
```

```python
v, minim = gd(f, df, -4, 0.1, 20)
colors = cm.hsv(np.linspace(0, 1, len(v)))
plt.plot(x,y)
plt.xlim(-4.3,4.3)
plt.ylim(-0.6,17)
for i, val in enumerate(v):
    plt.scatter(val, f(val), color=colors[i])
    if i == 0:
        plt.annotate(str(i+1), (val + 0.3, f(val)- 0.55), textcoords="offset points", xytext=(0,5), ha='center')
    if i <= 8 and i != 0:
        plt.annotate(str(i+1), (val, f(val)), textcoords="offset points", xytext=(0,5), ha='center')
    if i == 10:
        plt.annotate("...", (val, f(val)), textcoords="offset points", xytext=(0,5), ha='center')
# plt.scatter(v, f(v),color = colors, s = 70)
plt.scatter(v[0], f(v[0]), color = colors[0], s = 150)
plt.scatter(v[-1], f(v[-1]), color = colors[-1], s = 150)
plt.show()

plt.plot(x,y)
bb = np.mean(v)
plt.scatter(bb, f(bb), color = colors[0], s = 150)
plt.show()
```
